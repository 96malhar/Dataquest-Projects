# Pandas, Histograms

```python
cols=['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig=plt.figure(figsize=(5,23))
for r in range(4):
    ax=fig.add_subplot(4,1,r+1)
    ax=recent_grads[cols[r]].plot(kind='hist',rot=40)
    ax.set_xlabel(cols[r])
    
```

```python
cols=['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig=plt.figure(figsize=(5,23))
for r in range(4,8):
    ax=fig.add_subplot(4,1,r-3)
    ax=recent_grads[cols[r]].plot(kind='hist',rot=40)
    ax.set_xlabel(cols[r])
```

# Comparing Number Of Men and Women In Each Category Of Major

```python
import numpy as np
df = recent_grads.pivot_table(index = 'Major_category', values=['Men','Women'],aggfunc=np.sum)
df.plot.bar(x=df.index,y=['Men','Women'])
df
```
