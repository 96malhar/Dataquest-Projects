# Introduction To The Data
The aim of this notebook is to highlight the potential bias that certain sites have towards movie reviews

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

reviews = pd.read_csv('fandango_scores.csv')
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
norm_reviews = reviews[cols]
norm_reviews.head()
```

# Bar Plot for Average User Ratings of Avengers : Age Of Ultron

```python
ax = norm_reviews.iloc[0][1:].plot.bar()
ax.set_xlabel('Rating Source')
ax.set_ylabel('Average Rating')
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + 0.1, p.get_height() + 0.05))
```

# Horizontal Bar Plot for Average User Ratings of Avengers : Age Of Ultron

```python
ax1 = norm_reviews.iloc[0][1:].plot.barh()
ax1.set_ylabel('Rating Source')
ax1.set_xlabel('Average Rating')
ax1.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
```

# Creating a Scatter plot

```python
ax2 = norm_reviews.plot.scatter(x='Fandango_Ratingvalue', y='RT_user_norm')
ax2.set(xlabel="Fandango",ylabel="Rotten Tomatoes",ylim=(0,5),xlim=(0,5))
```

# Finding Correlation Between Fandango And Other Sites

```python
fig, axes = plt.subplots(3,1,figsize=(10,10))

col1 = ['RT_user_norm','Metacritic_user_nom','IMDB_norm']
col2 = ['Rotten Tomatoes','Metacritic','IMDB']

for i in range(3):
    norm_reviews.plot.scatter(x='Fandango_Ratingvalue',y=col1[i],ax=axes[i])
    axes[i].set(xlabel='Fandango',ylabel=col2[i],ylim=(0,5),xlim=(0,5))
```

# Generating And Commparing Histograms 

```python
fig,axis = plt.subplots(4,1,figsize=(15,35))

xt = np.arange(0,5.5,0.5)

col = ['Fandango_Ratingvalue','RT_user_norm','Metacritic_user_nom','IMDB_norm']
title = ['Distribution of Fandango Ratings',' Distribution of Rotten Tomatoes Ratings',
         'Distribution of Metacritic Ratings',' Distribution of IMDB Ratings']

for i in range(4):
    norm_reviews[col[i]].plot.hist(bins=20,ax=axis[i],title=title[i],ylim=(0,25),xlim=(0,5))
    axis[i].set_xticks(xt)
```

# Creating Box Plots to Visualize Data Distribution

```python
fig,ax = plt.subplots(figsize=(10,10))
norm_reviews.iloc[:][1:].plot.box(ax=ax,ylim=(0,5))
```
