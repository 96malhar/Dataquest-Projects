# Movie Reviews

```python
# import dataset
import pandas as pd
pd.set_option('display.max_columns', 100)
movies = pd.read_csv('fandango_score_comparison.csv')
movies
```

# Histograms

```python
import matplotlib.pyplot as plt
% matplotlib inline

# Histogram for Metacritic_norm_round
movies['Metacritic_norm_round'].plot.hist(title='Metacritic norm round')
plt.show()

# Histogram for Fandango_stars
movies['Fandango_Stars'].plot.hist(title='Fandango Stars')
plt.show()
```

## Fandango vs Metacritic Scores
1. Fandango_Stars values span over a smaller range with no values below 3.0
2. Metacritics_norm_round values span over a larger range starting from 0.5 to 4.5.

# Mean, median, and standard deviation

```python
# Descriptive scatistics for Metacritic_norm_round.
metacritic_mean = movies['Metacritic_norm_round'].mean()
metacritic_median = movies['Metacritic_norm_round'].median()
metacritic_stdev = movies['Metacritic_norm_round'].std()

# Descriptive scatistics for Fandango_stars.
fandango_mean = movies['Fandango_Stars'].mean()
fandango_median = movies['Fandango_Stars'].median()
fandango_stdev = movies['Fandango_Stars'].std()

print('metacritic_mean:',metacritic_mean)
print('metacritic_median:',metacritic_median)
print('metacritic_stdev:',metacritic_stdev)

print('fandango_mean:',fandango_mean)
print('fandango_median:',fandango_median)
print('fandango_stdev:',fandango_stdev)
```

## Metacritic Methodology
They curate a large number of reviews form reputed critics and summarize them using weighted averages(more importance is given to some critics). They also normalize the resulting scores to prevent them from clumping together. They do no inculde user votes in calculation of Metascores. More can be read [here](http://www.metacritic.com/about-metascores)

## Fandango Methodology
There is no transparent methodology to Fandango moview stars. However, it can be cleary observed that Fandango rates almost all its movies on a scale of 3-5. It includes user ratings in its stars. It also never rounds down the average user ratings and always rounds them up. Thus, a movie with an average rating of 4.1 will be rounded-off to 4.5. More about this can be read [here](https://fivethirtyeight.com/features/fandango-movies-ratings/)

## Fandango VS Metacritic numbers
The median Metacritic score is higher than the mean because a few lower values pull the mean to the left.

The median Fandango score is lower than the mean beacause a few very high values pull the mean to the right.

Metacritic scores have a higher standard deviation than Fandango scores because Metacritic scores span over a larger range of values which results in more variance from the mean.

# Correlations

```python
# Calculaing correlation between Metacritic and Fandango scores
from scipy.stats import pearsonr
r_value, p_value = pearsonr(movies['Metacritic_norm_round'], movies['Fandango_Stars'])
r_value
```

### Fandango and Metacritic correlation
The low correlation between Fandango and Metacritic scores indicates that Fandango scores aren't just inflated, they are fundamentally different. For whatever reason, it appears like Fandango both inflates scores overall, and inflates scores differently depending on the movie.

```python
# Creating Linear Regression

from scipy.stats import linregress

x = movies['Metacritic_norm_round']
y = movies['Fandango_Stars']

slope, intercept, r_value, p_value, stderr_slope = linregress(x ,y)
```

```python
# Predict what a movie that got a 3.0 in Metacritic would get on Fandango
pred_3 = 3 * slope + intercept
pred_3
```

# Finding residuals

```python
# Predict Fandango score with Metacritic values of 1 and 5.
pred_1 = 1 * slope + intercept
pred_5 = 5 * slope + intercept

# Make Scatter plot.
plt.scatter(movies["Metacritic_norm_round"], movies["Fandango_Stars"])

# make a line plot joining pred_1 and pred_5.
plt.plot([1,5],[pred_1,pred_5])
plt.xlim(1,5)
plt.show()
```
