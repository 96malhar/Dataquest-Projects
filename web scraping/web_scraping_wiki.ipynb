```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India")
content = response.content

parser = BeautifulSoup(content, 'html.parser')
```

```python
table = parser.find_all('table', class_='wikitable sortable plainrowheaders')[0]
```

```python
a = []
b = []
c = []
d = []
e = []
f = []
g = []

for row in table.find_all('tr')[1:]: #0th index contains the column names
    
    cells = row.find_all('td')
    states = row.find_all('th')
    
    if len(cells)==6: #data is flawed cell missing for punjab.
        
        a.append(cells[0].get_text())
        b.append(states[0].get_text())
        c.append(cells[1].get_text())
        d.append(cells[2].get_text())
        e.append(cells[3].get_text())
        f.append(cells[4].get_text())
        g.append(cells[5].get_text())
```

```python
import pandas as pd

data = {'number':a, 'state/ut':b, 'admin_capital':c, 
        'legislative_capital':d, 'judiciary_capital':e, 'year_capital':f, 'former_capital':g}

df = pd.DataFrame(data,columns=['number', 'state/ut', 'admin_capital',
                                'legislative_capital', 'judiciary_capital', 
                                'year_capital', 'former_capital'])
df
```
